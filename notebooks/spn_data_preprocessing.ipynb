```python
observation_rolling_72H = faults_diagnostics.groupby(['EquipmentID']).rolling('72H', closed = 'right').agg(agg_dict).reset_index().set_index('EventTimeStamp')
```

```python
prediction_rolling_24H = faults_diagnostics.loc[::-1].groupby(['EquipmentID'], as_index = False) \
    [['1569', '5246']] \
    .rolling('24H', closed = 'right').sum()[::-1].sort_values(['EquipmentID', 'EventTimeStamp']) \
    .rename(columns = {'1569': 'partial_derate_within_24hr', '5246': 'full_derate_within_24hr'}) 
prediction_rolling_24H.loc[prediction_rolling_24H['partial_derate_within_24hr'] > 0, 'partial_derate_within_24hr'] = 1
prediction_rolling_24H.loc[prediction_rolling_24H['full_derate_within_24hr'] > 0, 'full_derate_within_24hr'] = 1
prediction_rolling_24H['derate_within_24hr'] = prediction_rolling_24H[['partial_derate_within_24hr', 'full_derate_within_24hr']].max(axis = 1)

prediction_rolling_48H = faults_diagnostics.loc[::-1].groupby(['EquipmentID'], as_index = False) \
    [['1569', '5246']] \
    .rolling('48H', closed = 'right').sum()[::-1].sort_values(['EquipmentID', 'EventTimeStamp']) \
    .rename(columns = {'1569': 'partial_derate_within_48hr', '5246': 'full_derate_within_48hr'}) 
prediction_rolling_48H.loc[prediction_rolling_48H['partial_derate_within_48hr'] > 0, 'partial_derate_within_48hr'] = 1
prediction_rolling_48H.loc[prediction_rolling_48H['full_derate_within_48hr'] > 0, 'full_derate_within_48hr'] = 1
prediction_rolling_48H['derate_within_48hr'] = prediction_rolling_48H[['partial_derate_within_48hr', 'full_derate_within_48hr']].max(axis = 1)

prediction_rolling_72H = faults_diagnostics.loc[::-1].groupby(['EquipmentID'], as_index = False) \
    [['1569', '5246']] \
    .rolling('72H', closed = 'right').sum()[::-1].sort_values(['EquipmentID', 'EventTimeStamp']) \
    .rename(columns = {'1569': 'partial_derate_within_72hr', '5246': 'full_derate_within_72hr'}) 
prediction_rolling_72H.loc[prediction_rolling_72H['partial_derate_within_72hr'] > 0, 'partial_derate_within_72hr'] = 1
prediction_rolling_72H.loc[prediction_rolling_72H['full_derate_within_72hr'] > 0, 'full_derate_within_72hr'] = 1
prediction_rolling_72H['derate_within_72hr'] = prediction_rolling_72H[['partial_derate_within_72hr', 'full_derate_within_72hr']].max(axis = 1)
```

```python
observation_72H_prediction_24H_48H_72H = pd.concat([
    faults_diagnostics.sort_values(['EquipmentID', 'EventTimeStamp']),
    prediction_rolling_24H.drop(columns = 'EquipmentID'), 
    prediction_rolling_48H.drop(columns = 'EquipmentID'), 
    prediction_rolling_72H.drop(columns = 'EquipmentID'), 
    observation_rolling_72H.drop(columns = 'EquipmentID')], axis = 1) \
    .dropna(subset = '111')
```
